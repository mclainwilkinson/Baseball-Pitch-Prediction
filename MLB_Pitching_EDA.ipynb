## Thoughts on pitch types
I need to reduce the number of pitch types. Already given that out of 2.8m pitches thrown in 3 years, the 2 types were only thrown 9 times, for instance, are useless to include in the model. I need to completely remove these (at most) 18 at bats and potentially others with unclassifiable pitches. Let's see how many this would include.

```python
low_counts = ['AB', 'FA', 'UN', 'PO', 'FO', 'EP', 'KN', 'IN', 'SC']
extra = len(pitches[pitches['pitch_type'].isin(low_counts)]['ab_id'].unique())
print('{} unusable at bats is {:.2f}% of all at bats'.format(extra, extra / 740389 * 100))
```

0.98% is pretty negligible. We would still have 740,389 - 7283 = 733,106 sequences for training/testing, which is plenty for a deep learning model. We can subset at bats below.

```python
dropabs = pitches[pitches['pitch_type'].isin(low_counts)]['ab_id'].tolist()
modelatbats = atbats[~atbats['ab_id'].isin(dropabs)]
modelpitches = pitches[~pitches['ab_id'].isin(dropabs)]
print('new total # pitches: {}'.format(len(modelpitches['pitch_type'])))
```

```python
modelpitches.groupby('pitch_type').size().sort_values(ascending=False)
```

```python
pitch_conversions = {
    'CH': 'C',
    'CU': 'B',
    'KC': 'B',
    'SL': 'B',
    'FC': 'F',
    'FF': 'F',
    'FS': 'F',
    'FT': 'F',
    'SI': 'F',
}
```

```python
import matplotlib.pyplot as plt

modelpitches = modelpitches.dropna()
modelpitches['simple_type'] = modelpitches['pitch_type'].apply(lambda x: pitch_conversions[x])
modelpitches[['simple_type', 'pitch_type']].head()
```

```python
modelpitches.groupby('simple_type')['pitch_type'].count().sort_values(ascending=False)
```

```python
pitch_plot = modelpitches.groupby(['simple_type', 'pitch_type']).size().reset_index().pivot(columns='pitch_type',
                                                                                              index='simple_type',
                                                                                              values=0)
pitch_plot = pitch_plot / 2837473.0
plot = pitch_plot.plot.bar(stacked=True)
plt.title('pitch type distribution')
plt.xlabel('pitch type')
plt.ylabel('proportion')
plt.legend(loc='right', bbox_to_anchor=(1.2, 0.5))
plt.show()
```

```python
pitch_plot.sum(axis=1)
```

```python
pitchdef = {'CH':'Changeup',
            'CU':'Curveball',
            'FA':'Fastball',
            'EP':'Eephus',
            'FC':'Cutter',
            'FF':'Four-seam fastball',
            'FO':'Pitchout',
            'FS':'Splitter',
            'FT':'Two-seam fastball',
            'IN':'Intentional ball',
            'KC':'Knuckle curve',
            'KN':'Knuckleball',
            'PO':'Pitchout',
            'SC':'Screwball',
            'SI':'Sinker',
            'SL':'Slider',
            'UN':'Unknown',
            'AB':'Unknown2'}

pitch_count = pitches.groupby('pitch_type').size().sort_values(ascending=False)
pitch_count.index = pitch_count.index.map(lambda x: pitchdef[x])
pitch_count.plot.barh()
plt.title('Pitch Count Types')
plt.ylabel('pitch type')
plt.xlabel('count')
plt.show()
```
